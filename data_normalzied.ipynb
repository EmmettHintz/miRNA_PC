```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv('./data/source_data/GSE211692_processed.txt', sep='\t')

# Display summary statistics
print(df.describe())

# Plot histograms for a few miRNA columns to check the distribution
miRNA_columns = df.columns[1:6]  # Adjust the range to include relevant miRNA columns

for col in miRNA_columns:
    plt.figure()
    df[col].hist(bins=50)
    plt.title(f'Histogram of {col}')
    plt.xlabel('Expression Value')
    plt.ylabel('Frequency')
    plt.show()
```
