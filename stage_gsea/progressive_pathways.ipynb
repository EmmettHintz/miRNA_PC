```python
import pandas as pd
from upsetplot import generate_data, UpSet, from_contents
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ctl_s1 = pd.read_csv('./ctl_s1/ctl_s1_genes.csv')
s1_s2 = pd.read_csv('./s1_s2/s1_s2_genes.csv')
s2_s3 = pd.read_csv('./s2_s3/s2_s3_genes.csv')
s3_s4 = pd.read_csv('./s3_s4/s3_s4_genes.csv')
```

```python
# Extract 'Gene Symbol' From each list
genes_ctl_s1 = ctl_s1['Gene Symbol']
genes_s1_s2 = s1_s2['Gene Symbol']
genes_s2_s3 = s2_s3['Gene Symbol']
genes_s3_s4 = s3_s4['Gene Symbol']

print(f'Genes in ctl_s1: {" ".join(genes_ctl_s1)}\n')
print(f'Genes in s1_s2: {" ".join(genes_s1_s2)}\n')
print(f'Genes in s2_s3: {" ".join(genes_s2_s3)}\n')
print(f'Genes in s3_s4: {" ".join(genes_s3_s4)}\n')
```

```python
ctl_s1 = pd.ExcelFile('./genecards_comparisons/ctl_s1.xlsx')
s1_s2 = pd.ExcelFile('./genecards_comparisons/s1_s2.xlsx')
s2_s3 = pd.ExcelFile('./genecards_comparisons/s2_s3.xlsx')
s3_s4 = pd.ExcelFile('./genecards_comparisons/s3_s4.xlsx')
```

```python
ctl_s1_pathways = pd.read_excel(ctl_s1, 'Pathways', skiprows=18)
s1_s2_pathways = pd.read_excel(s1_s2, 'Pathways', skiprows=18)
s2_s3_pathways = pd.read_excel(s2_s3, 'Pathways', skiprows=18)
s3_s4_pathways = pd.read_excel(s3_s4, 'Pathways', skiprows=18)

ctl_s1_diseases = pd.read_excel(ctl_s1, 'Diseases', skiprows=18)
s1_s2_diseases = pd.read_excel(s1_s2, 'Diseases', skiprows=18)
s2_s3_diseases = pd.read_excel(s2_s3, 'Diseases', skiprows=18)
s3_s4_diseases = pd.read_excel(s3_s4, 'Diseases', skiprows=18)
```

```python
# print(f'ctl_s1_pathways columns: {ctl_s1_pathways.columns}\n')
# print(f's1_s2_pathways columns: {s1_s2_pathways.columns}\n')
# print(f's2_s3_pathways columns: {s2_s3_pathways.columns}\n')
# print(f's3_s4_pathways columns: {s3_s4_pathways.columns}\n')

# print(f'ctl_s1_diseases columns: {ctl_s1_diseases.columns}\n')
# print(f's1_s2_diseases columns: {s1_s2_diseases.columns}\n')
# print(f's2_s3_diseases columns: {s2_s3_diseases.columns}\n')
# print(f's3_s4_diseases columns: {s3_s4_diseases.columns}\n')
```

```python
ctl_s1_pathways_set = ctl_s1_pathways[['SuperPath Name', 'Score']].rename(columns={'Score': 'ctl_s1'})
s1_s2_pathways_set = s1_s2_pathways[['SuperPath Name', 'Score']].rename(columns={'Score': 's1_s2'})
s2_s3_pathways_set = s2_s3_pathways[['SuperPath Name', 'Score']].rename(columns={'Score': 's2_s3'})
s3_s4_pathways_set = s3_s4_pathways[['SuperPath Name', 'Score']].rename(columns={'Score': 's3_s4'})
```

```python
#  Merge the data on 'SuperPath Name'
merged_df = ctl_s1_pathways_set
merged_df = merged_df.merge(s1_s2_pathways_set, on='SuperPath Name', how='outer')
merged_df = merged_df.merge(s2_s3_pathways_set, on='SuperPath Name', how='outer')
merged_df = merged_df.merge(s3_s4_pathways_set, on='SuperPath Name', how='outer')



# Calculate the score changes only for the numeric columns
numeric_cols = merged_df.columns.drop('SuperPath Name')
merged_df['score_change'] = merged_df[numeric_cols].max(axis=1) - merged_df[numeric_cols].min(axis=1)

# Sort by score changes
sorted_df = merged_df.sort_values(by='score_change', ascending=False)

# Select top pathways with the most significant changes
top_changes_df = sorted_df.head(10)

# Plot the scores of pathways that change the most over time
plt.figure(figsize=(20, 8))
stages = numeric_cols

for index, row in top_changes_df.iterrows():
    plt.plot(stages, row[numeric_cols], marker='o', label=row['SuperPath Name'])

plt.xlabel('Stages of Cancer Progression')
plt.ylabel('Scores')
plt.title('Top 10 Pathways with Most Significant Score Changes Over Time')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# Display the top pathways with their score changes
print(f'Top pathways with most significant score changes:\n{top_changes_df}')
```

```python
# Filter pathways that are present in all groups
pathways_in_all_groups = merged_df.dropna()

# Calculate the average score for each pathway across all groups (numeric columns only)
numeric_cols_pathways = pathways_in_all_groups.columns.drop('SuperPath Name')
pathways_in_all_groups['average_score'] = pathways_in_all_groups[numeric_cols_pathways].mean(axis=1)

# Sort pathways by the average score
sorted_pathways = pathways_in_all_groups.sort_values(by='average_score', ascending=False)

# Select top pathways with the highest average score
top_pathways = sorted_pathways.head(10)

# Print the top pathways with their average scores
print("Top Pathways Involved Across All Groups (Highest Average Scores):")
print(top_pathways[['SuperPath Name', 'average_score']])

# Create the UpSet plot for pathways
data_pathways = {
    'ctl_s1': set(ctl_s1_pathways['SuperPath Name']),
    's1_s2': set(s1_s2_pathways['SuperPath Name']),
    's2_s3': set(s2_s3_pathways['SuperPath Name']),
    's3_s4': set(s3_s4_pathways['SuperPath Name'])
}

upset_data_pathways = from_contents(data_pathways)
upset_pathways = UpSet(upset_data_pathways, subset_size='count')
upset_pathways.plot()
```

```python
# Extract 'Name' and 'Score' columns
ctl_s1_diseases_set = ctl_s1_diseases[['Name', 'Score']].rename(columns={'Score': 'ctl_s1'})
s1_s2_diseases_set = s1_s2_diseases[['Name', 'Score']].rename(columns={'Score': 's1_s2'})
s2_s3_diseases_set = s2_s3_diseases[['Name', 'Score']].rename(columns={'Score': 's2_s3'})
s3_s4_diseases_set = s3_s4_diseases[['Name', 'Score']].rename(columns={'Score': 's3_s4'})

# Merge the data on 'Name'
merged_df = ctl_s1_diseases_set.merge(s1_s2_diseases_set, on='Name', how='outer')
merged_df = merged_df.merge(s2_s3_diseases_set, on='Name', how='outer')
merged_df = merged_df.merge(s3_s4_diseases_set, on='Name', how='outer')

# Filter diseases that are present in all groups
diseases_in_all_groups = merged_df.dropna()

# Calculate the average score for each disease across all groups (numeric columns only)
numeric_cols = diseases_in_all_groups.columns.drop('Name')
diseases_in_all_groups['average_score'] = diseases_in_all_groups[numeric_cols].mean(axis=1)

# Sort diseases by the average score
sorted_diseases = diseases_in_all_groups.sort_values(by='average_score', ascending=False)

# Select top diseases with the highest average score
top_diseases = sorted_diseases.head(10)

# Print the top diseases with their average scores
print("Top Diseases Involved Across All Groups (Highest Average Scores):")
print(top_diseases[['Name', 'average_score']])

# Create the UpSet plot for diseases
data_diseases = {
    'ctl_s1': set(ctl_s1_diseases['Name']),
    's1_s2': set(s1_s2_diseases['Name']),
    's2_s3': set(s2_s3_diseases['Name']),
    's3_s4': set(s3_s4_diseases['Name']),
}

upset_data_diseases = from_contents(data_diseases)
upset_diseases = UpSet(upset_data_diseases, subset_size='count')
upset_diseases.plot()

# Show the plot
plt.show()
```
