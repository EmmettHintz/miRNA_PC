```python
stage_row = df.iloc[12]
print(stage_row)
```

```python
column_names = df.iloc[0].values

disease_df = pd.DataFrame([disease_row.values[1:]], columns=column_names[1:], index=['Disease']).T
sex_df = pd.DataFrame([sex_row.values[1:]], columns=column_names[1:], index=['Sex']).T
age_df = pd.DataFrame([age_row.values[1:]], columns=column_names[1:], index=['Age']).T
stage_df = pd.DataFrame([stage_row.values[1:]], columns=column_names[1:], index=['Stage']).T
```

```python
metadata_df = pd.concat([sex_df, age_df, stage_df, disease_df], axis=1)
metadata_df.head()
```

```python
# Load the processed data
df = pd.read_csv('../data/GSE211692.txt', sep='\t', header=0, index_col=0)
df.head()
```

```python
id_ref = df.index
print(id_ref)
```

```python
metadata_df.index.name = 'ID_REF'
df_with_metadata = df.T.join(metadata_df)

print("Merged DataFrame:")
print(df_with_metadata.head())
```
