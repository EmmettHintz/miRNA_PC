```python
import pandas as pd
```

```python
ctl_s1 = pd.read_csv('./miRNA/50_ctl_s1.csv')
s1_s2 = pd.read_csv('./miRNA/50_s1_s2.csv')
s2_s3 = pd.read_csv('./miRNA/50_s2_s3.csv')
s3_s4 = pd.read_csv('./miRNA/50_s3_s4.csv')
```

```python
# sort each df by 'Importance' score
ctl_s1 = ctl_s1.sort_values(by='Importance', ascending=False)
s1_s2 = s1_s2.sort_values(by='Importance', ascending=False)
s2_s3 = s2_s3.sort_values(by='Importance', ascending=False)
s3_s4 = s3_s4.sort_values(by='Importance', ascending=False)
```

```python
# print each row of the df on a new line
print('CTL_S1 miRNA & IMPORTANCE SCORES')
print(ctl_s1.to_string(index=False), '\n')

print('S1_S2 miRNA & Importance Scores')
print(s1_s2.to_string(index=False), '\n')

print('S2_S3 miRNA & Importance Scores')
print(s2_s3.to_string(index=False), '\n')

print('S3_S4 miRNA & Importance Scores')
print(s3_s4.to_string(index=False), '\n')
```
