```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
s1_s2_miRNAs = pd.read_csv('./miRNA/s1_s2_miRNA.csv')
s2_s3_miRNAs = pd.read_csv('./miRNA/s2_s3_miRNA.csv')
s3_s4_miRNAs = pd.read_csv('./miRNA/s3_s4_miRNA.csv')
```

```python
# make a new dataframe that has the columns of s1_s2_miRNAs and s2_s3_miRNAs and s3_s4_miRNAs
merged = pd.concat([s1_s2_miRNAs, s2_s3_miRNAs, s3_s4_miRNAs], axis=1)
merged.head()
```

```python
# Flatten the dataframe into a single series and count the occurrences
miRNA_counts = merged.stack().value_counts().sort_values(ascending=False)

# Print the miRNA counts
print(miRNA_counts)

# Plot the top 10 most frequent miRNAs
top_miRNAs = miRNA_counts.head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_miRNAs.values, y=top_miRNAs.index)
plt.xlabel('Count')
plt.ylabel('miRNA')
plt.title('Top 10 most frequent miRNAs')
plt.show()
```

```python
from matplotlib_venn import venn3

# Creating sets of miRNAs
s1_s2_set = set(s1_s2_miRNAs.stack())
s2_s3_set = set(s2_s3_miRNAs.stack())
s3_s4_set = set(s3_s4_miRNAs.stack())

# Plotting the Venn diagram
plt.figure(figsize=(8, 8))
venn3([s1_s2_set, s2_s3_set, s3_s4_set], ('s1 vs s2', 's2 vs s3', 's3 vs s4'))
plt.title('Venn Diagram of miRNA Intersections')
plt.show()
```
